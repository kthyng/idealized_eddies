# Idealized Eddy Simulations

Rob ran a series of 30 day idealized shelf simulations with the density set up to possibly generate eddies that could be similar to those seen in the summer on the Texas shelf. I ran surface drifters in the simulations to help characterize and compare with the real eddies.

One drifter was initialized in each grid cell daily at the surface and run for the remainder of the simulation.



## Look at timing of eddy development using energy

Rob has a script in his idealized eddy simulations directory `energy.py` for calculating the total, mean, and eddy kinetic energy. I want to use these calculations to look at the timing of the eddy development in stages and compare between different runs. I copied the file and moved it into a similar directory of mine on merrimack so I could change it to run through all the simulations and make a plot (`kmt_shelfstrat/energy.py`). 

The plot is set up analogously to another plot of the parameter space, Ri vs. S (slope burger number). The highlighted bottom 6 plots have at least some measure of eddy kinetic energy in the system, and those are the ones I will do analysis on. Grey arrows and text indicate the approximate start and stop times of the ramp up timing of the eddies, according to the eddy kinetic energy in the system, and the 'f' number indicates the frame number of that time in the frames.
